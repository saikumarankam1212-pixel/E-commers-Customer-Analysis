### **Explanaton & Visualization**

#### **Gender Analysis**

```python
print(df['Gender'].value_counts())
```

```python
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Analysis")
```

###### **Customer Mostly Equal males and femals**

### **Total Revenue by Product Category**

```python
category_revenue = df.groupby("Product Category")["Total Purchase Amount"].sum().sort_values(ascending=False)

print(category_revenue)
```

```python
import matplotlib.pyplot as plt

category_revenue.plot(kind="bar", figsize=(8,5))

plt.title("Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

###### **Electronics generated the highest revenue**
###### **Books generated the lowest revenue.**

### **Average Product Price by Category**

```python
AVG_product_category=df.groupby("Product Category")["Product Price"].mean()
print(AVG_product_category)
```

```python
AVG_product_category.plot(
    kind='bar'
)

plt.title("Total Revenue by Product Category")

plt.show()
```

###### **Almost each Category Product Avg price 250**

### **Average Purchase Amount by Gender**

```python
avg_purchase = df.groupby("Gender")["Total Purchase Amount"].mean()

print(avg_purchase)
```

```python
avg_purchase.plot(kind="bar")

plt.title("Average Purchase Amount by Gender")
plt.ylabel("Average Purchase")
plt.show()
```

###### **Female customers spent more per transaction than male customers.**

### **Number of Customers by Gender**

```python
Gender_no_of_customer=df.groupby("Gender")["Customer ID"].count()
print(Gender_no_of_customer)
```

```python
Gender_no_of_customer.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("No of Customers Gender")

plt.show()
```

### **Total Revenue by Gender**

### **Total Revenue by Payment Method**

### **Customer Age Distribution**

### **A noticeable percentage of customers have churned.**

```python
top_customers = df.groupby("Customer Name")["Total Purchase Amount"].sum().sort_values(ascending=False).head(10)

print(top_customers)
```

```python
top_customers.plot(kind="barh")

plt.title("Top 10 Customers")
plt.xlabel("Revenue")
plt.show()
```

###### **A small number of customers contribute a large portion of total revenue.**

```python
correlation = df[["Age","Quantity","Product Price","Total Purchase Amount"]].corr()

print(correlation)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))

plt.imshow(correlation, cmap="coolwarm")

plt.colorbar()

plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=45)
plt.yticks(range(len(correlation.columns)), correlation.columns)

plt.title("Correlation Heatmap")

plt.show()
```

###### **Product Price shows a positive relationship with Total Purchase Amount.**
###### **Quantity also influences total revenue.**
